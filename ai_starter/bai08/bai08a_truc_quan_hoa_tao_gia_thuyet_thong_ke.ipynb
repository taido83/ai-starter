```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Tải tập dữ liệu
Ames = pd.read_csv('../data/Ames.csv')

# Tính hệ số tương quan của tất cả các đặc trưng với 'SalePrice'
correlations = Ames.corr(numeric_only=True)['SalePrice'].sort_values(ascending=False)

# Lấy 5 đặc trưng tương quan cao nhất với 'SalePrice' (loại bỏ chính 'SalePrice')
top_5_features = correlations.index[1:6]

# Tạo biểu đồ pair plot cho các đặc trưng này và 'SalePrice'
# Điều chỉnh kích thước bằng cách thiết lập chiều cao và tỷ lệ
sns.pairplot(Ames, vars=['SalePrice'] + list(top_5_features), height=1.35, aspect=1.85)

# Hiển thị biểu đồ
plt.show()
```

```python
# Chuyển đổi 'LotShape' thành biến nhị phân: 'Regular' và 'Irregular'
Ames['LotShape_Binary'] = \
    Ames['LotShape'].apply(lambda x: 'Regular' if x == 'Reg' else 'Irregular')

# Tạo biểu đồ pair plot, tô màu theo biến 'LotShape_Binary'
sns.pairplot(Ames, vars=['SalePrice', 'OverallQual', 'GrLivArea'], hue='LotShape_Binary',
             palette='Set1', height=2.5, aspect=1.75)

# Hiển thị biểu đồ
plt.show()
```

```python
# Tạo pair plot cho 5 đặc trưng tương quan nhất với SalePrice
sns.pairplot(Ames, vars=['SalePrice', 'GrLivArea', 'GarageCars', 'TotalBsmtSF', '1stFlrSF', 'OverallQual'],
             height=1.35, aspect=1.85)
```
